```python
import os
import shutil
import json
import subprocess

path = r"G:\HK2N4\BOBSL_Lightweight\sign_language_web\sign_language_web\frontend"

try:
    os.chdir(path)
    print(f"Đang xử lý tại: {os.getcwd()}\n")

    # Bước 1: Xóa cache cũ bị lỗi của gh-pages
    cache_path = os.path.join("node_modules", ".cache", "gh-pages")
    if os.path.exists(cache_path):
        print("1. Đang xóa bộ nhớ đệm (cache) bị lỗi của gh-pages...")
        shutil.rmtree(cache_path, ignore_errors=True)
        print("-> Đã xóa thành công.")
    else:
        print("1. Không tìm thấy thư mục cache bị lỗi, bỏ qua.")

    # Bước 2: Sửa package.json để đổi -r thành --repo
    print("\n2. Đang kiểm tra và sửa package.json...")
    with open("package.json", "r", encoding="utf-8") as f:
        pkg = json.load(f)

    if "scripts" in pkg and "deploy" in pkg["scripts"]:
        old_cmd = pkg["scripts"]["deploy"]
        new_cmd = old_cmd.replace("-r ", "--repo ")
        if old_cmd != new_cmd:
            pkg["scripts"]["deploy"] = new_cmd
            with open("package.json", "w", encoding="utf-8") as f:
                json.dump(pkg, f, indent=2)
            print("-> Đã đổi '-r' thành '--repo' trong script deploy.")
        else:
            print("-> Script deploy đã đúng định dạng.")

    # Bước 3: Ép cấu hình lại Git Remote origin (Đảm bảo an toàn)
    print("\n3. Đang thiết lập lại git remote...")
    subprocess.run(["git", "remote", "remove", "origin"],
                   stderr=subprocess.DEVNULL)
    subprocess.run(["git", "remote", "add", "origin",
                   "https://github.com/n22dccn196/signbridge.git"], check=True)
    print("-> Đã thiết lập thành công remote origin.")

    # Bước 4: Chạy lệnh deploy cuối cùng
    print("\n4. Đang tiến hành deploy (Có thể mất 15-30 giây)...")
    result = subprocess.run(["npm", "run", "deploy"],
                            shell=True, text=True, capture_output=True)

    print("\n--- KẾT QUẢ DEPLOY ---")
    print(result.stdout)
    if result.stderr:
        print("\n--- THÔNG BÁO / LỖI ---")
        print(result.stderr)
        if "Published" in result.stderr or "Published" in result.stdout:
            print("\n=> THÀNH CÔNG! Tuy có thông báo ở dải stderr nhưng code đã được push lên nhánh gh-pages của GitHub!")

except Exception as e:
    print(f"Lỗi hệ thống trong quá trình chạy: {e}")
```

```python
import os
import subprocess
import shutil

# Đường dẫn thư mục frontend
frontend_path = r"G:\HK2N4\BOBSL_Lightweight\sign_language_web\sign_language_web\frontend"
repo_url = "https://github.com/n22dccn196/signbridge.git"

try:
    os.chdir(frontend_path)
    print(f"Đang làm việc tại: {os.getcwd()}\n")

    # 1. Chạy Build để đảm bảo code mới nhất
    print("1. Đang build project (npm run build)...")
    build = subprocess.run("npm run build", shell=True,
                           capture_output=True, text=True)
    if build.returncode != 0:
        print("LỖI BUILD:\n", build.stderr)
        exit(1)
    print("-> Build thành công.")

    # 2. Di chuyển vào thư mục dist
    dist_path = os.path.join(frontend_path, "dist")
    if not os.path.exists(dist_path):
        print(f"\nLỖI: Không tìm thấy thư mục {dist_path}")
        exit(1)

    os.chdir(dist_path)
    print(f"\n2. Đã di chuyển vào thư mục build: {os.getcwd()}")

    # 3. Xóa thư mục .git trong dist (nếu có từ lần chạy trước) để làm mới
    if os.path.exists(".git"):
        shutil.rmtree(".git", ignore_errors=True)

    # 4. Thực hiện chuỗi lệnh Git để đẩy thẳng lên branch gh-pages
    print("\n3. Đang khởi tạo Git và đẩy thẳng code lên GitHub (Bỏ qua thư viện gh-pages)...")

    commands = [
        "git init",
        "git add -A",
        'git commit -m "Deploy to gh-pages"',
        "git branch -M gh-pages",
        f"git push -f {repo_url} gh-pages"
    ]

    for cmd in commands:
        print(f"-> Chạy: {cmd}")
        res = subprocess.run(cmd, shell=True, capture_output=True, text=True)
        # Bỏ qua lỗi rỗng của commit (nếu không có file nào mới)
        if res.returncode != 0 and "commit" not in cmd:
            print(f"LỖI:\n{res.stderr}")
            exit(1)

    print("\n=> THÀNH CÔNG RỰC RỠ! Đã đẩy trực tiếp thư mục dist lên GitHub nhánh gh-pages!")
    print("Bạn có thể truy cập GitHub > Settings > Pages để xem trang web của mình.")

except Exception as e:
    print(f"\nLỗi hệ thống: {e}")
```

```python
import os

# Thư mục gốc chứa cả frontend và backend
base_path = r"G:\HK2N4\BOBSL_Lightweight\sign_language_web\sign_language_web"

print("--- 1. TÌM VỊ TRÍ THỰC SỰ CỦA package.json ---")
for root, dirs, files in os.walk(base_path):
    if "node_modules" in root:
        continue
    if "package.json" in files:
        print(f"-> Đã thấy package.json tại: {root}")

print("\n--- 2. TÌM FILE CHỨA LOCALHOST TRONG SOURCE CODE ---")
for root, dirs, files in os.walk(base_path):
    # Bỏ qua thư mục build và thư viện
    if "node_modules" in root or "dist" in root or ".git" in root:
        continue
    for file in files:
        if file.endswith((".js", ".jsx", ".ts", ".tsx", ".env")):
            filepath = os.path.join(root, file)
            try:
                with open(filepath, 'r', encoding='utf-8') as f:
                    content = f.read()
                    if 'localhost' in content or '127.0.0.1' in content:
                        print(f"-> Phát hiện 'localhost' tại file: {filepath}")
            except Exception:
                pass
```

```python
import os
import subprocess
import shutil

# =====================================================================
# DÁN LINK NGROK BẠN VỪA COPY VÀO ĐÂY
# Ví dụ: PUBLIC_API_URL = "https://1234-abcd.ngrok-free.app"
PUBLIC_API_URL = "DÁN_LINK_NGROK_VÀO_ĐÂY"
# =====================================================================

project_path = r"G:\HK2N4\BOBSL_Lightweight\sign_language_web\sign_language_web"
repo_url = "https://github.com/n22dccn196/signbridge.git"

try:
    if PUBLIC_API_URL == "DÁN_LINK_NGROK_VÀO_ĐÂY":
        print("LỖI: Bạn chưa thay đổi biến PUBLIC_API_URL. Hãy dán link ngrok vào script và chạy lại.")
        exit(1)

    os.chdir(project_path)
    print(f"Đang làm việc tại: {os.getcwd()}\n")

    print(f"1. Đang build project với VITE_API_BASE = {PUBLIC_API_URL} ...")
    env = os.environ.copy()
    env["VITE_API_BASE"] = PUBLIC_API_URL

    build = subprocess.run("npm run build", shell=True,
                           env=env, capture_output=True, text=True)
    if build.returncode != 0:
        print("LỖI BUILD:\n", build.stderr)
        exit(1)
    print("-> Build thành công với URL API mới.")

    # Tìm thư mục dist (nằm trong thư mục gốc hoặc trong frontend)
    dist_path = os.path.join(project_path, "frontend", "dist")
    if not os.path.exists(dist_path):
        dist_path = os.path.join(project_path, "dist")
        if not os.path.exists(dist_path):
            print(f"LỖI: Không tìm thấy thư mục dist sau khi build.")
            exit(1)

    os.chdir(dist_path)
    print(f"\n2. Đã di chuyển vào thư mục build: {os.getcwd()}")

    if os.path.exists(".git"):
        shutil.rmtree(".git", ignore_errors=True)

    print("\n3. Đang đẩy code mới lên GitHub Pages...")
    commands = [
        "git init",
        "git add -A",
        f'git commit -m "Deploy with ngrok URL: {PUBLIC_API_URL}"',
        "git branch -M gh-pages",
        f"git push -f {repo_url} gh-pages"
    ]

    for cmd in commands:
        res = subprocess.run(cmd, shell=True, capture_output=True, text=True)
        if res.returncode != 0 and "commit" not in cmd:
            print(f"LỖI TẠI LỆNH '{cmd}':\n{res.stderr}")
            exit(1)

    print("\n=> THÀNH CÔNG RỰC RỠ! Frontend đã được cập nhật URL backend mới và đẩy lên GitHub!")
    print("LƯU Ý: Mỗi lần bạn tắt ngrok và bật lại, link sẽ thay đổi và bạn sẽ cần chạy lại script này với link mới.")

except Exception as e:
    print(f"\nLỗi hệ thống: {e}")
```
